## PaySim Preprocessing and Graph Construction

The PaySim dataset contains synthetic mobile money transactions designed to simulate realistic financial behavior and fraud patterns. The goal is to prepare both tabular representations for baseline machine learning models and graph-structured data for Graph Neural Networks (GNNs).

We perform log transformation on transaction amounts, encode categorical transaction types numerically, and standardize selected numerical features to ensure consistent scaling across models.

For graph construction, each transaction is represented as a node. Instead of constructing a dense similarity graph, we build a temporally constrained transaction-flow graph. Edges are created based on shared account participation and short-range temporal proximity, capturing meaningful sequential behavior in account activity while avoiding over-connection between unrelated transactions. This design allows the graph to reflect realistic transaction propagation patterns and burst-like fraudulent activity commonly observed in financial systems.

```python
# Feature engineering
df_paysim = df_paysim.copy()

df_paysim['log_amount'] = np.log1p(df_paysim['amount'])
df_paysim['type_encoded'] = LabelEncoder().fit_transform(df_paysim['type'])

features_p = [
    'log_amount','type_encoded','oldbalanceOrg',
    'newbalanceOrig','oldbalanceDest','newbalanceDest','step'
]

X_paysim = df_paysim[features_p]
y_paysim = df_paysim['isFraud']

scaler = StandardScaler()
X_paysim_scaled = scaler.fit_transform(X_paysim)

df_paysim['tx_id'] = np.arange(len(df_paysim))



# Graph construction (Temporal + Flow)
edges = []

# sort by time
df_paysim = df_paysim.sort_values("step")

orig_map = df_paysim.groupby("nameOrig")["tx_id"].apply(list)
dest_map = df_paysim.groupby("nameDest")["tx_id"].apply(list)

# Flow edges (same account sequences)
for _, txs in orig_map.items():
    if len(txs) > 1:
        txs = sorted(txs)
        edges += list(zip(txs[:-1], txs[1:]))

for _, txs in dest_map.items():
    if len(txs) > 1:
        txs = sorted(txs)
        edges += list(zip(txs[:-1], txs[1:]))

# Temporal window edges
WINDOW = 3

for acc, g in df_paysim.groupby("nameOrig"):
    txs = g["tx_id"].values
    steps = g["step"].values

    for i in range(len(txs)):
        for j in range(i+1, len(txs)):
            if steps[j] - steps[i] <= WINDOW:
                edges.append((txs[i], txs[j]))
            else:
                break

edges_paysim = pd.DataFrame(edges, columns=["source", "target"]).drop_duplicates()


# Node Table
nodes_paysim = pd.DataFrame(X_paysim_scaled, columns=features_p)
nodes_paysim["tx_id"] = df_paysim["tx_id"].values
nodes_paysim["label"] = y_paysim.values


# cleanup
del X_paysim, X_paysim_scaled
gc.collect()

print(f"Nodes shape: {nodes_paysim.shape}")
print(f"Edges shape: {edges_paysim.shape}")
```

## IEEE-CIS Preprocessing and Graph Construction

The IEEE-CIS dataset consists of online transaction records enriched with associated identity information such as card details, email domains, and device metadata. We first merge transaction and identity tables, then handle missing values by imputing numerical features with medians and filling categorical variables with an "unknown" token prior to label encoding.

For baseline machine learning models, all features are standardized to ensure consistent scaling across heterogeneous inputs.

For graph construction, each transaction is represented as a node. Instead of relying on artificial ordering or sequential assumptions, edges are constructed using shared high-signal entity relationships. Transactions are connected when they share common attributes such as card identifiers, email domains, or device-related fields. This results in a hub-style interaction graph that captures behavioral similarity through shared entities, enabling the model to learn coordinated fraud patterns across related transactions.

```python
def preprocess_ieee(df_trans, df_id, is_train=True):

    df_ieee = df_trans.merge(df_id, on="TransactionID", how="left")
    df_ieee = df_ieee.copy()

    # Missing value handling
    num_cols = df_ieee.select_dtypes(include=np.number).columns
    for col in num_cols:
        df_ieee[col] = df_ieee[col].fillna(df_ieee[col].median())

    cat_cols = df_ieee.select_dtypes(include="object").columns
    for col in cat_cols:
        df_ieee[col] = df_ieee[col].fillna("unknown")
        df_ieee[col] = LabelEncoder().fit_transform(df_ieee[col].astype(str))

    # Label/features
    if is_train:
        y = df_ieee["isFraud"].values
        df_ieee.drop(columns=["TransactionID", "isFraud"], inplace=True)
    else:
        y = None
        df_ieee.drop(columns=["TransactionID"], inplace=True)

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(df_ieee)

    df_ieee["tx_id"] = np.arange(len(df_ieee))

    # Graph constructin (hub-based)
    edges = []

    key_cols = ["card1", "P_emaildomain", "R_emaildomain", "DeviceType"]

    for col in key_cols:
        if col not in df_ieee.columns:
            continue

        grouped = df_ieee.groupby(col)["tx_id"].apply(list)

        for _, txs in grouped.items():
            if len(txs) > 1:
                hub = txs[0]
                for t in txs[1:]:
                    edges.append((hub, t))
                    edges.append((t, hub))

    edges = pd.DataFrame(edges, columns=["source", "target"]).drop_duplicates()

    # Node table
    nodes = pd.DataFrame(X_scaled, columns=df_ieee.columns.drop("tx_id"))
    nodes["tx_id"] = df_ieee["tx_id"].values

    if is_train:
        nodes["label"] = y


    del df_ieee, X_scaled
    gc.collect()

    return nodes, edges


nodes_ieee_train, edges_ieee_train = preprocess_ieee(df_ieee_trans, df_ieee_id, True)
nodes_ieee_test, edges_ieee_test = preprocess_ieee(df_ieee_test_trans, df_ieee_test_id, False)

print(nodes_ieee_train.shape, edges_ieee_train.shape)
print(nodes_ieee_test.shape, edges_ieee_test.shape)
```

## Elliptic Preprocessing and Graph Construction

The Elliptic dataset is inherently graph-structured and represents Bitcoin transactions as nodes with directed edges capturing the flow of funds between transactions. We first rename feature columns for clarity, merge transaction features with class labels, and remove transactions with unknown labels to ensure supervised training consistency. We also make sure illicit transactions are labeled with 1, and licit transactions are labeled with :0.

Node features consist of precomputed transaction-level and neighborhood-level attributes that describe both local transaction properties and aggregated structural information from the surrounding network.

Edges are directly taken from the provided transaction flow edgelist, preserving the original directed Bitcoin transaction graph. This structure naturally encodes the propagation of funds through the network, enabling the model to learn patterns associated with illicit transaction chains and laundering behavior.

Each node corresponds to a unique transaction with its feature vector and class label (licit vs illicit), maintaining full graph integrity for GNN-based learning.

```python
num_cols = df_elliptic_features.shape[1]
feature_cols = [f"feature_{i}" for i in range(1, num_cols-1)]
df_elliptic_features.columns = ["txId", "timestep"] + feature_cols


df_elliptic = df_elliptic_features.merge(df_elliptic_classes, on="txId")
df_elliptic = df_elliptic[df_elliptic["class"] != "unknown"].copy()
df_elliptic["class"] = df_elliptic["class"].astype(int)
df_elliptic["class"] = df_elliptic["class"].replace({2: 0})

X_elliptic = df_elliptic.drop(columns=["txId", "class"])
y_elliptic = df_elliptic["class"]

scaler = StandardScaler()
X_elliptic_scaled = scaler.fit_transform(X_elliptic)

nodes_elliptic = pd.DataFrame(X_elliptic_scaled, columns=X_elliptic.columns)
nodes_elliptic["tx_id"] = df_elliptic["txId"].values
nodes_elliptic["label"] = y_elliptic.values


edges_elliptic = df_elliptic_edges.copy()
edges_elliptic.columns = ["source", "target"]

# Remov Self Loops
edges_elliptic = edges_elliptic[edges_elliptic["source"] != edges_elliptic["target"]]


del df_elliptic_features, df_elliptic_classes, df_elliptic
gc.collect()

print(nodes_elliptic.shape)
print(edges_elliptic.shape)
```

## Conclusion

In this preprocessing pipeline, we transform three diverse financial transaction datasets into structured formats suitable for both traditional machine learning and graph-based learning approaches. Through careful preprocessing—including normalization, encoding, and missing value handling—we construct robust tabular representations for baseline models.

In parallel, we construct graph representations that reflect meaningful relational structure within each dataset. Rather than relying on arbitrary similarity connections, we design graphs based on temporal proximity, shared entity relationships, and true transactional flows. This ensures that the resulting graphs capture realistic dependencies relevant to fraud behavior.

The resulting node and edge datasets provide a reproducible foundation for evaluating both classical machine learning models and Graph Neural Networks, enabling systematic comparison of feature-based versus structure-aware fraud detection approaches.

## Use for Baseline Machine Learning Models

The node datasets contain standardized feature vectors and corresponding labels, making them directly suitable for classical machine learning models such as Logistic Regression, K-Nearest Neighbors, Decision Trees, and Random Forests. Each row represents an individual transaction with engineered features that summarize its behavior in isolation.

These models provide strong baselines for evaluating predictive performance and help identify which individual transaction attributes carry the most discriminative signal for fraud detection.

## Use for Graph Neural Network Models

The combined node and edge datasets form graph representations suitable for Graph Neural Network architectures such as GCN, GAT, and GraphSAGE. Nodes represent individual transactions with associated feature vectors, while edges encode meaningful relationships such as shared entities, temporal proximity, or transactional fund flow.

These graph structures enable models to learn from both feature-level information and relational dependencies between transactions. As a result, GNNs can capture higher-order patterns such as coordinated fraud behavior, repeated account usage, and multi-transaction fraud chains that are not easily detectable using traditional tabular methods.
